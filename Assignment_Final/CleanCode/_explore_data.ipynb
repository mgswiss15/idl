```python
import torch
```

```python
dset = 'liver_data.pt'
pt_path='../data/'+dset
data_dict = torch.load(pt_path)
```

```python
# Inside your data creation script
# Merging Train and Val into one bulk training block
data_dict["train_images"] = torch.cat((data_dict["train_images"], data_dict["val_images"]), dim=0)
data_dict["train_labels"] = torch.cat((data_dict["train_labels"], data_dict["val_labels"]), dim=0)
del data_dict["val_images"]
del data_dict["val_labels"]

data_dict.keys()
```

```python
torch.save(data_dict, pt_path)
```
