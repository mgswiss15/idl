```python
import torch
```

```python
X = torch.tensor([[1., 2., 1.],
                  [0., 1., 3.],
                  [2., 0., 1.],
                  [1., 1., 2.]])          # (4, 3)

Y = torch.tensor([[1., 0.],
                  [0., 1.],
                  [1., 1.],
                  [0., 0.]])              # (4, 2)

W = torch.tensor([[1., 0., 1.],
                  [0., 1., 0.]])          # (2, 3)

b = torch.tensor([0., 0.])               # (2,)
```

```python
print(f"shape of X {X.shape}")
n = X.shape[0]
```

```python
# residuals
R = X @ W.T + b - Y
```

```python
# Gradients
grad_W = 2/n * R.T @ X
grad_b = 2/n * R.sum(dim=0)
print(f"grad_W {grad_W}, {grad_W.shape}")
print(f"grad_b {grad_b}, {grad_b.shape}")
```
