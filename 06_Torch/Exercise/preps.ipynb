Prepare data for MLP exercise

This is a quick an dirty data prep for an in-class competition of Intro to DL within MAI. 

Original dataset: https://archive.ics.uci.edu/dataset/206/relative+location+of+ct+slices+on+axial+axis

Dropping patients for simplicity though this is a major data leakage. 

```python
import pandas as pd
```

```python
# load data from csv
df = pd.read_csv('slice_localization_data.csv')
df
```

```python
# drop patient id
df_nop = df.drop(columns='patientId').rename(columns={'reference': 'TARGET'})
df_nop
```

```python
# prepar for split 80/20
df_nop = df_nop.sample(frac=1).reset_index(drop=True)
N = len(df_nop)
0.8 * N
```

```python
# split to train test
train = df_nop[:42800]
train
```

```python
test = df_nop[42800:]
test
```

```python
# export to csv files
train.to_csv('ct_slice_train.csv', index_label='ID')
test.to_csv('ct_slice_test.csv', index_label='ID')
test.drop(columns='TARGET').to_csv('ct_slice_test_noLabels.csv', index_label='ID')
tg = test['TARGET']
tg_df = pd.DataFrame(tg)
tg_df['Usage'] = 'Public'
tg_df['Usage'].iloc[5350:] = 'Private'
tg_df.to_csv('ct_slice_test_Labels.csv', index_label='ID')
tg.values[:]=0
tg.to_csv('sample_submission.csv', index_label='ID')
tg_df
```

```python
# sandbox submission
avg = train['TARGET'].mean()
tg.values[:]=avg
tg.to_csv('sandbox.csv', index_label='ID')
```
